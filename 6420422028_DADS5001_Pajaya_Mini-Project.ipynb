# การใช้ไฟฟ้าของทั้งประเทศ (จำแนกตาม Sector)

```python
# อ่านไฟล์ชื่อการใช้ไฟฟ้าของทั้งประเทศ (จำแนกตามสาขา)
Electricity_Usage = pd.read_csv('การใช้ไฟฟ้าของทั้งประเทศ (จำแนกตามสาขา).csv')
Electricity_Usage
```

```python
df_groupby = Electricity_Usage.groupby(['Year','Month','Sector'])[['Quantity']].agg('mean')
df_groupby
```

```python
df_1 = Electricity_Usage.pivot_table( index='Year', columns='Sector', values='Quantity', aggfunc='mean' )
df_1
```

```python
# เทียบปริมาณการใช้ไฟฟ้าของทั้งประเทศ (จำแนกตามสาขา) ในแต่ละปี
sns.set(rc={'figure.figsize':(10,5)})
ax = sns.heatmap(df_1, linewidths=0.5,cmap="YlOrRd")
```

```python
df_merge = pd.merge( ft, df_1, left_on='Year', right_on='Year', how='left' )
df_merge
```

```python
df_merge['Sum Sectors'] = (df_merge['Agriculture'] + df_merge['Business'] + df_merge['Free of Charge'] + df_merge['Government & Non-Profit']+ df_merge['Industrial']+ df_merge['Other']+ df_merge['Residential'] )
df_merge
```

```python
df_merge2 = df_merge.drop( columns=['Agriculture','Business','Free of Charge','Government & Non-Profit', 'Industrial', 'Other','Residential'] ) 
df_merge2
```

```python
# Plot graph ปริมาณการใช้ไฟฟ้าของทั้งประเทศ (จำแนกตามสาขา) และค่า Ft ในแต่ละปี โดยอ้างอิงเดือน ม.ค. เป็นตัวแทนของเดือนช่วงปลายปี (ม.ค.-เม.ย.)
sns.relplot(x="Sum Sectors", y="Jan", hue="Year", 
            sizes=(10, 100), alpha=.5, palette="muted",
            height=6, data=df_merge2)
```

```python
# Plot graph ปริมาณการใช้ไฟฟ้าของทั้งประเทศ (จำแนกตามสาขา) และค่า Ft ในแต่ละปี โดยอ้างอิงเดือน ต.ค. เป็นตัวแทนของเดือนช่วงปลายปี (ต.ค.-ธ.ค.)
sns.relplot(x="Sum Sectors", y="Oct", hue="Year", 
            sizes=(10, 100), alpha=.5, palette="muted",
            height=6, data=df_merge2)
```
